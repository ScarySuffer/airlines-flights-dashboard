```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='airline', y='price', data=df)
plt.title('💸 Flight Price Distribution by Airline')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Determine 99th percentile price
price_cap = df['price'].quantile(0.99)

# Step 2: Filter out outliers
df_clean = df[df['price'] <= price_cap]

print(f"🧼 Cleaned data shape: {df_clean.shape}")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='airline', y='price', data=df_clean)
plt.title('💸 Flight Price Distribution by Airline (No Outliers)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='class', y='price', data=df_clean, estimator='mean')
plt.title('🎟️ Average Flight Price by Class')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='stops', y='price', data=df_clean, estimator='mean')
plt.title('🛑 Average Price by Number of Stops')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='departure_time', y='price', data=df_clean, estimator='mean', order=[
    'Early_Morning', 'Morning', 'Afternoon', 'Evening', 'Night', 'Late_Night'
])
plt.title('🕒 Avg Price by Departure Time')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='days_left', y='price', data=df_clean)
plt.title('📉 Flight Price vs Days Left')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Price (ZAR)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_clean.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('📊 Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
df_clean['route'] = df_clean['source_city'] + ' → ' + df_clean['destination_city']
top_routes = df_clean.groupby('route')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_routes.values, y=top_routes.index, palette='viridis')
plt.title('🌆 Top 10 Expensive Routes (Avg Price)')
plt.xlabel('Average Price')
plt.tight_layout()
plt.show()
```
